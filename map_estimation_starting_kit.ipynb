## Introduction

The continuous monitoring of blood pressure is a major challenge in the general anesthesia field. Indeed, the monitoring of blood pressure is essential to ensure that the patient is stable during the operation. However, the current methods to measure blood pressure are either non-invasive, but non-continuous, or invasive, which can lead to complications, and expensive, making them inadapted in many situations.

<div style="text-align: center;">
<img src="https://github.com/ramp-kits/map_estimation/raw/main/img/cuff.avif" width=350px alt="Cuff for blood pressure measurment"></img>
</div>

A potential solution to this problem is to use non-invasive monitoring signals which are routinely collected, like the electrocardiogram (ECG) photoplethysmogram (PPG) signal, to estimate the mean arterial pressure (MAP) using AI. The PPG signal is a non-invasive signal that can be easily acquired using a pulse oximeter. The MAP is a measure of the average blood pressure in an individual's arteries during one cardiac cycle.

<div style="text-align: center;">
<img src="https://github.com/ramp-kits/map_estimation/raw/main/img/ecg.avif" height=250px alt="ECG sensors"></img>
<img src="https://github.com/ramp-kits/map_estimation/raw/main/img/ppg.png" height=250px alt="PPG sensor"></img>
</div>

The goal of this challenge is to estimate the MAP from the non-invasive signals. The dataset consists of window of 10 seconds of PPG and ECG signals, with the related patient information (age, gender, weight, height, etc.) and the MAP measured with an invasive method.

A particular aspect of this challenge is the fact that we would like to have a
method that can generalize well to new patients in other hospitals. This is a
major issue in the medical field, where the data distribution can change and
it is very costly to collect new data.

For this reason, the challenge is a domain adaptation challenge. For training,
you have access to a dataset composed of data from 2 different domains:

- The "source" domain `'v'`, which corresponds to data collected in a first hospital.
  For this data, you have access to all the data and the labels (i.e. the MAP).
- The "target" domain `'m'`, composed of data collected in a second hospital.
  For this data, you have access to the data, but not the labels. In order to make it possible to test that everything is working properly, we give you the labels for 100 segments of the target domain in the test set.

The goal is therefore to generalize from the source domain to the target domain.

# Challenge evaluation

For this challenge, the metric used to evaluate the performance of the model is the mean absolute error (MAE) between the predicted MAP and the true MAP, defined as:

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_{pred}^i - y_{true}^i|
$$

Here, we take the MAE over the test set, which is composed of signals coming
from domain `m`.

## The pipeline workflow

The input data are stored in a dataframe. To go from a dataframe to a numpy array we will use a scikit-learn column transformer. The first example we will write will just consist in selecting a subset of columns we want to work with.


```python
# %load submissions/starting_kit/estimator.py

from sklearn import set_config
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OrdinalEncoder

set_config(transform_output="pandas")


class IgnoreDomain(RandomForestRegressor):
    def fit(self, X, y):
        # Ignore the samples with missing target
        X = X[y != -1]
        y = y[y != -1]
        return super().fit(X, y)


def get_estimator():
    return make_pipeline(
        make_column_transformer(
            ("passthrough", ["age"]),
            (OrdinalEncoder(
                handle_unknown='use_encoded_value', unknown_value=-1
            ), ["gender"]),
        ),
        IgnoreDomain(n_estimators=50)
    )
```

## CNN :

L'objectif est d'extraire les features de la même manière pour les 2 time series dont on dipose.
Pour cela, une approche CNN est peut-être la plus pertinente au première abord :

L'architecture est simple :

- Input les données de ECG et PPG (1250 par segment)
- Conv1D
- ReLU
- Max/Average pooling
- Conv1D
- ReLU
- Max/Average pooling

--> Concatenation des 2 embedding extraits puis utilisation evaluation en utilisant un mlp classique


```python
import torch
import torch.nn as nn
```

```python
class CNN1D(nn.Module):
    def __init__(self, 
                 conv_kernel1,
                 conv_kernel2,
                 nb_filters,
                 pool_kernel1,
                 pool_kernel2,
                 stride,
                 pooling = nn.MaxPool1d
    ):
        super().__init__()

        self.conv1 = nn.Conv1d(in_channels=1,
                               out_channels=nb_filters,
                               kernel_size=conv_kernel1,
                               stride = stride,
                               )
        self.activation1 = nn.ReLU()
        self.pool1 = pooling(kernel_size = pool_kernel1, stride = stride)
        self.conv2 = nn.Conv1d(in_channels=nb_filters,
                               out_channels=nb_filters,
                               kernel_size=conv_kernel2,
                               stride = stride,
                               )
        self.activation2 = nn.ReLU()
        self.pool2 = pooling(kernel_size = pool_kernel2, stride = stride)

        torch.nn.init.xavier_normal_(self.conv1.weight)
        torch.nn.init.xavier_normal_(self.conv2.weight)
        
        self.extractor = nn.Sequential(
            self.conv1,
            self.activation1,
            self.pool1,
            self.conv2,
            self.activation2,
            self.pool2
        )

    def forward(self,x):
        return self.extractor(x)
    
```

```python
# Setup of the data : 
test_data = torch.Tensor(X_df.iloc[0]["ecg"])
test_data.size(), test_data
```

```python
class ModelCNN(nn.Module) :
    def __init__(self, 
                 CNN_ECG_settings,
                 CNN_PPG_settings,
                 size_output # tuple : (nb_filter, output_size)
                 ):

        super().__init__()
        self.ecg_extractor = CNN1D(**CNN_ECG_settings)
        self.ppg_extractor = CNN1D(**CNN_PPG_settings)

        flattened_size = size_output[0] * size_output[1] 
        print(flattened_size)

        self.norm_ecg = nn.BatchNorm1d(num_features=size_output[0])
        self.norm_ppg = nn.BatchNorm1d(num_features=size_output[0])

        self.lin1 = nn.Linear(in_features = flattened_size * 2, out_features = 1000)
        self.activation1 = nn.ReLU()
        self.lin2 = nn.Linear(in_features = 1000, out_features = 200)
        self.activation2 = nn.ReLU()
        self.lin3 = nn.Linear(in_features = 200, out_features = 1)

        self.regressor = nn.Sequential(
            self.lin1, 
            self.activation1,
            self.lin2,
            self.activation2,
            self.lin3
        )

    def forward(self, ecg_data, ppg_data) :
        # x should have this structure : (ecg, ppg)
        ecg = ecg_data.unsqueeze(0)
        emb_ecg = self.ecg_extractor(ecg)

        ppg = ppg_data.unsqueeze(0)
        emb_ppg = self.ppg_extractor(ppg)

        norm_ecg = self.norm_ecg(emb_ecg)
        norm_ppg = self.norm_ppg(emb_ppg)

        norm_ecg = norm_ecg.view(-1)
        norm_ppg = norm_ppg.view(-1)

        emb_concat = torch.concat((norm_ecg, norm_ppg))

        return self.regressor(emb_concat)
```

```python
settings_ecg = {
    "conv_kernel1" : 6,
    "conv_kernel2" : 5,
    "nb_filters" : 10,
    "pool_kernel1" : 6,
    "pool_kernel2" : 4,
    "stride" :  2,
    "pooling" : nn.MaxPool1d
}

settings_ppg = {
    "conv_kernel1" : 6,
    "conv_kernel2" : 5,
    "nb_filters" : 10,
    "pool_kernel1" : 6,
    "pool_kernel2" : 4,
    "stride" :  2,
    "pooling" : nn.MaxPool1d
}

size_output = (10, 75)
```

```python
from torch.utils.data import Dataset,DataLoader


def convert_map_to_tensor(data):
    num_data = data.shape[0]
    np_array = np.ndarray((num_data, 1250))
    for ind, val in enumerate(data): 
        np_array[ind, :] = val
    return torch.from_numpy(np_array).float()



class MapDataset(Dataset):
    def __init__(self,
                 num_source,
                 num_target,
                 train = True,
                 source_only = True):
        if train : 
            X_v, y_v = problem.get_train_data(start=0, stop=num_source)
            X_m, y_m = problem.get_train_data(start=-(num_target+1), stop=-1)
            X_df = pd.concat([X_v, X_m], axis=0)
            y = np.concatenate([y_v, y_m])

            if source_only : 
                df, target  = X_v, y_v
            else :
                df, target = X_df,y
        else : 
            df, target = problem.get_test_data()

        self.ecg_data = convert_map_to_tensor(df["ecg"])
        self.ppg_data = convert_map_to_tensor(df["ppg"])
        self.map = torch.from_numpy(target.reshape(-1, 1)).float()
    
    def __len__(self):
        return self.ecg_data.size(0)
    
    def __getitem__(self, index):
        return self.ecg_data[index], self.ppg_data[index], self.map[index]
```

```python
Map_train = MapDataset(num_source=1000, num_target=1000)
Map_test = MapDataset(num_source=1000, num_target=1000, train=False)
```

```python
mod_pred = ModelCNN(CNN_ECG_settings= settings_ecg, CNN_PPG_settings=settings_ppg, size_output=size_output)
```

```python
train_loader = DataLoader(dataset=Map_train,batch_size=1, shuffle=True)
test_loader = DataLoader(dataset=Map_test,batch_size=1, shuffle=True)
```

```python
from tqdm import tqdm

EPOCHS = 30
LR = 1E-3

loss = nn.L1Loss()
optim = torch.optim.SGD(mod_pred.parameters(), lr=LR)

for epoch in tqdm(range(EPOCHS)):
    for ecg_data, ppg_data, target_map in train_loader:
        optim.zero_grad()
        y_pred = mod_pred(ecg_data,ppg_data)
        l = loss(y_pred, target_map)
        l.backward()
        optim.step()
        
```

```python
cum_loss = 0
for ecg_test, ppg_test, y_test in test_loader: 
    with torch.no_grad():
        pred = mod_pred(ecg_test, ppg_test)
        l = loss(pred, y_test)
        cum_loss += l
cum_loss / len(test_loader)
```
